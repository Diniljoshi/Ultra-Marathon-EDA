### Question 2 What age group(count > 20 ) are the best and worst in 50k and 50 mi event?

### Question 3 What is the average speed of athletes in different season?

```python
new_df['event_season'] = new_df['event_month'].apply(lambda x:'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
new_df
```

```python
new_df.groupby(['event_season'])['athlete_avg_speed'].agg(['mean','count']).sort_values(by = 'mean',ascending = False) 
```

```python
#50 mi only
new_df.query('event_distance == "50mi"').groupby(['event_season'])['athlete_avg_speed'].agg(['mean','count']).sort_values(by = 'mean',ascending = False) 
```

```python
#50 km only
new_df.query('event_distance == "50km"').groupby(['event_season'])['athlete_avg_speed'].agg(['mean','count']).sort_values(by = 'mean',ascending = False) 
```

```python
#Seasonwise both for female and male
new_df.groupby(['event_season', 'athlete_gender'])['athlete_avg_speed'].agg(['mean', 'count']).reset_index().pivot_table(index='event_season', columns='athlete_gender', values=['mean', 'count'], aggfunc='sum')

```
